```python
# Import Packages

import numpy as np
import pandas as pd

import numpy as np

import nltk

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# import os
# for dirpath, dirnames, filenames in os.walk('.'):
#     for f in filenames:
#         print(os.path.join(dirpath, f))
```

```python
# Import Data - and tokenise

train_data = pd.read_csv("./nlp-getting-started/train.csv")

text_train = train_data['text'].values
targets_train = train_data['target'].values

n_vals = len(text_train)

tokenised = np.array(np.zeros(n_vals), dtype=object)

for i in range(n_vals):
    tokenised[i] = word_tokenize(text_train[i])
```
