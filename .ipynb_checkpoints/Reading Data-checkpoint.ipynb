```python
# Import Packages

import numpy as np
import pandas as pd

import numpy as np

import nltk
# may need to download packages -> uncomment below to do so

# nltk.download('averaged_perceptron_tagger')
# nltk.download('wordnet')
# nltk.download('stopwords')

# nltk pre-processing packages
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk import WordNetLemmatizer
from nltk.corpus import stopwords


# for finding directories

# import os
# for dirpath, dirnames, filenames in os.walk('.'):
#     for f in filenames:
#         print(os.path.join(dirpath, f))
```

```python
# Import Data

train_data = pd.read_csv("../nlp-getting-started/train.csv")

text_train = train_data['text'].values
targets_train = train_data['target'].values
```

# Tokenisation

```python
#  Tokenised Data

n_vals = len(text_train)

tokenised = np.array(np.zeros(n_vals), dtype=object)

for i in range(n_vals):
    tokenised[i] = word_tokenize(text_train[i])

```

```python
# Printing Tokenised
index = 13
print(text_train[index])
print(tokenised[index],targets_train[index])
```

# Tagging

```python
# Tagged words

tagged = np.array(np.zeros(n_vals), dtype=object)
for i in range(n_vals):
    tagged[i] = nltk.pos_tag(tokenised[i])
```

```python
# Printing Tagged
index =13
print(text_train[index])
print(tagged[index])
```

# Lemmatisation

```python
lemma = nltk.WordNetLemmatizer()

lemmatised = np.array(np.zeros(n_vals), dtype=object)

for i in range(n_vals):
    text = tokenised[i]
    ex = [i.lower() for i in text]
    lemmatised[i] = [lemma.lemmatize(i) for i in ex]
```

```python
# Printing Lemmatised
index =687
print(text_train[index])
print(lemmatised[index])
```

# Stemming

```python
lancaster = nltk.LancasterStemmer()
porter = nltk.PorterStemmer()

lancaster_stems = np.array(np.zeros(n_vals), dtype=object)
porter_stems = np.array(np.zeros(n_vals), dtype=object)

for i in range(n_vals):
    tokens = tokenised[i]
    lancaster_stems[i] = [porter.stem(i) for i in tokens]
    porter_stems[i] = [lancaster.stem(i) for i in tokens]
```

```python
# Printing Stemmed
index = 12
print(tokenised[index])
print(lancaster_stems[index])
print(porter_stems[index])
```

# Word Stopping

```python
stop_words = set(stopwords.words('english'))
stop_filtered = np.array(np.zeros(n_vals), dtype=object)

for i in range(n_vals):
    filtered = []
    for word in tokenised[i]: 
        if word not in stop_words: 
            filtered.append(word)
    stop_filtered[i] = filtered    
```

```python
# Printing word stopped
index = 125
print(tokenised[index])
print(stop_filtered[index])
```
